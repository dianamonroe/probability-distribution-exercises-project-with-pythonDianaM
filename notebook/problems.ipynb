## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)
    
    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Write the above functions but for both normal and uniform distributions.

```python
# EXERCISE 1. 1) NORMAL DISTRIBUTION FUNCTIONS

from scipy.stats import norm


def pdfnormal(x, loc, scale):
    """
    Calculates the probability density function
    """
    result = norm.pdf(x, loc, scale)

    return result

def cdfnorm(x, loc, scale): 
    """
    Calculates the cumulative of the normal distribution
    """
    result = norm.cdf(x, loc, scale)

    return result

def qauntinorm(q, loc, scale):
    """
    Calculates the quantile function from the normal distribution
    """
    result = norm.ppf(q, loc, scale)

    return result

def rnorm(loc, scale, size):
    """
    Generates random variables from the normal distribution
    """
    result = norm.rvs(loc, scale, size)

    return result
```

```python
# Normal(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Normal(10, 0.2) takes the value 2: {pdfnormal(x = 2, loc = 10, scale = 0.2)}")
print(f"Probability that a Normal(10, 0.2) will take a value less than 2: {cdfnorm(x = 2, loc = 10, scale = 0.2)}")
print(f"Which value of a Normal(10, 0.2) has a cumulative probability of 0.9? {qauntinorm(q = 0.9, loc = 10, scale = 0.2)}")
print(f"Generate 2000 random values from a Normal(10, 0.2) distribution: {rnorm(loc=0, scale=1, size=2000)}") 
```

```python
# PLOT EXERCISE 1.1) NORMAL DISTRIBUTION FUNCTIONS
import matplotlib.pyplot as plt


np.random.seed(42)

size = 2000
loc = 0  
scale = 1  
data = np.random.normal(loc=loc, scale=scale, size=size) 

pdf = pdfnormal(data, loc = loc, scale = scale)
cdf = cdfnorm(data, loc = loc, scale = scale)
ppf = qauntinorm(data, loc = size, scale = scale)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

# Plot 1: Probability Density Function (PDF)
axis[0].stem(data, pdf, "r-")
axis[0].set_title("Normal Probability Density Function (PDF)")
axis[0].set_xlabel("Normal Random Variable (x)")
axis[0].set_ylabel("Normal Probability Density")

# Plot 2: Cumulative Distribution Function (CDF)
axis[1].stem(data, cdf, "b-")
axis[1].set_title("Normal Cumulative Distribution Function (CDF)")
axis[1].set_xlabel("Normal Random Variable (x)")
axis[1].set_ylabel("Normal Cumulative Probability")

# Plot 3: Percent Point Function (PPF) or Quantile Function
axis[2].stem(data, ppf, "g-")
axis[2].set_title("Normal Percent Point Function (PPF)")
axis[2].set_xlabel("Normal Probability (p)")
axis[2].set_ylabel("Normal Quantile (x)")

plt.tight_layout()
plt.legend()

plt.show()
```

```python
# EXERCISE 1. 2) UNIFORM DISTRIBUTION FUNCTIONS

from scipy.stats import uniform


def pdfuniform(x, loc=0, scale=1):
    """
    Calculates the probability density function of the uniform distribution
    """
    result = uniform.pdf(x=x, loc=loc, scale=scale) 

    return result

def cdfuniform(q, loc=0, scale=1):
    """
    Calculates the cumulative of the uniform distribution
    """
    result = uniform.cdf(x=q, loc=loc, scale=scale) 

    return result

def qauntinuniform(p, loc=0, scale=1):

    """
    Calculates the quantile function from the uniform distribution
    """
    result = uniform.ppf(q=p, loc=loc, scale=scale) 

    return result

def runiform(n, loc=0, scale=1):
    """
    Generates random variables from the uniform distribution
    """
    result = uniform.rvs(size=n, loc=loc, scale=scale)  

    return result
```

```python
# PRINTS EX. 1.1. Uniform(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Uniform(10, 0.2) takes the value 2: {pdfuniform(2, loc = 1, scale = 1)}")
print(f"Probability that a Uniform(10, 0.2) will take a value less than 2: {cdfuniform(2, loc = 1, scale = 0.2)}") 
print(f"Which value of a Uniform(10, 0.2) has a cumulative probability of 0.9? {qauntinuniform(0.9, loc=10, scale=0.2)}") 
print(f"Generate 2000 random values from a Uniform(10, 0.2) distribution: {runiform(n = 2000, loc=0, scale=1)}")
```

```python
# PLOT EXERCISE 1.2) UNIFORM DISTRIBUTION FUNCTIONS
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform

np.random.seed(42)

size = 2000
loc = 0  
scale = 1  
x = runiform(size, loc, scale)  
x.sort()

pdf = pdfuniform(x, loc, scale)
cdf = cdfuniform(x, loc, scale)
ppf = qauntinuniform(np.linspace(loc, scale, size), loc, scale)

fig, axis = plt.subplots(3, 1, figsize=(5, 7))

# Plot 1: Probability Density Function (PDF)
axis[0].plot(x, pdf, "r-") 
axis[0].set_title("Uniform Probability Density Function (PDF)")
axis[0].set_xlabel("Uniform Random Variable (x)")
axis[0].set_ylabel("Uniform Probability Density")

# Plot 2: Cumulative Distribution Function (CDF)
axis[0].fill_between(x, pdf, color='red', alpha=0.3)
axis[1].plot(x, cdf, "b-") 
axis[1].set_title("Uniform Cumulative Distribution Function (CDF)")
axis[1].set_xlabel("Uniform Random Variable (x)")
axis[1].set_ylabel("Uniform Cumulative Probability")

# Plot 3: Percent Point Function (PPF) or Quantile Function
axis[2].plot(x, ppf, "g-") 
axis[2].set_title("Uniform Percent Point Function (PPF)")
axis[2].set_xlabel("Uniform Probability (p)")
axis[2].set_ylabel("Uniform Quantile (x)")

plt.tight_layout()
plt.show()
```
